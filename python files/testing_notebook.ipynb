```python
table_list = df_dic['FST20030321']
```

```python
new_list = []
i = 0
for table in table_list:
    i += 1
    
    ### PRINT INFORMATION ABOUT THE TABLE
    print("@"*50, "\n", "Processing table: " + str(i))
    display(HTML(table.to_html()))
    
    # DATE CLEANER
    table = dateCleaner(table, key_words = "", stop_words = "")
    
    # VOLUME CLEANER
    table = volumeCleaner(table, key_words = "", stop_words = "")
    
    #PRICE CLEANER
    table = priceCleaner(table, key_words = "", stop_words = "")
    
    table.columns = table.columns.str.lower()
    unprocessed = columnSorter(table)
    print("Unprocessed columns: ", "\n", unprocessed)
    display(HTML(table.to_html()))
    new_list.append(table)
```
